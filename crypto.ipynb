```python
import requests as r
import matplotlib.pyplot as plt
import datetime as dt
price = []
time = []
def getDatafromURL(fromSymbol,toSymbol,limit):
    url = "https://min-api.cryptocompare.com/data/histominute?" \
          "fsym="+fromSymbol.upper()+"&tsym="+toSymbol.upper()+"&limit="+limit+"&aggregate=1&e=CCCAGG"
    req = r.get(url)
    dataSet = req.json()['Data']
    for data in dataSet:
        time.append(dt.datetime.fromtimestamp(float(data['time'])))
        price.append(float(data['close']))

def plotting(fromSymbol,toSymbol,limit):
    getDatafromURL(fromSymbol,toSymbol,limit)
    plt.plot(time, price)
    plt.title(fromSymbol.upper()+' convert to '+toSymbol.upper())
    plt.xlabel('Time in minute')
    plt.ylabel('Price in '+toSymbol.upper())
    plt.show()


plotting('eth','usd','9999')
plotting('btc','usd','9999')
```
